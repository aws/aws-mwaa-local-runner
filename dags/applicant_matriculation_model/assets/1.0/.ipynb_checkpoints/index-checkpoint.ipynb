#### Web Scraping - Programming Languages & Frameworks

```python
def get_survey_data(year):
    """
    The yearly survey data can be downloaded as a zipfile
    which contains a `survey_results_public.csv` file.
    
    """
    url = f'https://info.stackoverflowsolutions.com/rs/719-EMH-566/images/stack-overflow-developer-survey-{year}.zip'
    content = get(url).content
    survey_zip = BytesIO(content)
    zipfile = ZipFile(survey_zip)
    csv = zipfile.open('survey_results_public.csv').read().decode('utf-8')
    so = pd.read_csv(StringIO(csv))
    return so
```

**Collecting the csv files for `2019-2021`**

```python
so_2021 = get_survey_data(2021)
so_2020 = get_survey_data(2020)
so_2019 = get_survey_data(2019)
```

**A few cleaning steps:**
- The surveys are not identical across all years.  Some surveys use a `LanguageDesireNextYear` column to report the "most popular language"
and others use `LanguageWantToWorkWith`. I'll need to check which of these two options are available and then isolate the column
- The data contained in the language column are string-lists for each row, containing all of the languages a used selected. Each language is seperated by a `;` character. 
- `bash/shell/powershell` is the option made available by stack overflow, but for our purposes, we will want each of those words isolated (Applicants are unlikely to describe bash usage in the same format.)

```python
languages = set()
for frame in [so_2021, so_2020, so_2019]:
    column_options = ['LanguageDesireNextYear', 'LanguageWantToWorkWith']
    column = [x for x in column_options if x in frame.columns][0]
    language_list = (frame[column].str.lower()
                                  .str.split(';') # Convert string-list to list
                                  .explode()
                                  .to_frame('languages')
                                  .dropna()
                                  .query('~languages.str.contains("other")')
                                  .languages.str.split('/') # Splitting 'bash/shell/powershell'
                                  .explode()
                                  .to_frame('languages')
                                  .languages.unique()
                                  .tolist())
    languages.update(language_list)
```

```python
se_syllabus_terms = ['react', 'api', 'rack', 'rails', 'json', 'git', 'github'] 
ds_syllabus_terms = ['machine learning', 'matplotlib',
                     'web scraping', 'visualization', 'data',
                     'statistics', 'analysis', 'pandas',
                     'numpy', 'scipy', 'statmodels', 'github', 'git',
                     'api', 'json']
cse_syllabus_terms = ['probability', 'ethernet', 'wifi', 'tcp', 
                     'network', 'systems', 'threat', 'metasploit',
                     'linux', 'cyberspace', 'governance', 'risk management',
                     'compliance', 'workstation', 'server', 'firewall',
                     'python', 'penetration', 'cryptography', 'ssl', 'tls',
                     'ssh', 'encrypt'
                 ]

pro_syllabus_terms = ['user', 'interface', 'design', 'typography',
                     'pattern', 'mockup', 'visual', 'research', 
                     'communicate', 'ethics', 'portfolio',
                     'process']

languages.update(se_syllabus_terms)
languages.update(ds_syllabus_terms)
languages.update(cse_syllabus_terms)
languages.update(pro_syllabus_terms)
```

I will be using regex to identify the presence of each key term in the `languages` dataset. Below I define a helper function to generate the regex pattern for a list of key terms. 

```python
def compile_key_term_pattern(terms):
    """
    terms - An iterable of key terms
    
    returns - A compiled regex pattern for identifing the presents of each term.
              The pattern is primarily just joining the iterable together with the
              `|` character
    """
    template = r'\b({terms})\b'
    pattern = template.format(terms='|'.join(([re.escape(x) for x in terms])))
    return re.compile(pattern)
```

Test the regex pattern:

```python
# Compile regex pattern
language_pattern = compile_key_term_pattern(languages)
# Run the pattern on an test string
language_pattern.findall('i know python, css, javascript, html, matplotlib')
```

#### Web Scraping - Free Learning Resources

I will collect a list of free learning resources from a `learntocodewith.me` article, that lists 100+ resources as headers.

```python
def get_resources():
    response = get('https://learntocodewith.me/posts/code-for-free/')
    soup = BeautifulSoup(response.text, 'lxml')
    h3 = soup.find_all('h3')
    resources = [x.find('a').text.lower() for x in h3 if x.find('a') and x.find('a').text]
    return resources
```

```python
resources = get_resources() + ['free code camp']
print(len(resources), 'learning resources collected.')
```

Testing the compiled regex for learning resources

```python
resources_pattern = compile_key_term_pattern(resources)
resources_pattern.findall('I am learning to code on coursera, udemy, codewars, and free code camp')
```

For this project, I will count the number of languages and learning resources mentioned.

Below, I define a function that
- Runs the compiled regex
- Counts the number of returned matches.

```python
def count_key_terms(text, terms):
    pattern = compile_key_term_pattern(terms)
    matches = set(pattern.findall(text))
    return len(matches)
```

### Generate the features

```python
df = df.assign(
         
         free_resources_mentioned=(df.technical_background.str.lower().fillna('').apply(count_key_terms, terms=resources) 
                                    + df.professional_background.str.lower().fillna('').apply(count_key_terms, terms=resources)),
         coding_languages_mentioned=(df.technical_background.str.lower().fillna('').apply(count_key_terms, terms=languages)
                                    + df.professional_background.str.lower().fillna('').apply(count_key_terms, terms=languages)),
         similarity=df[['technical_background', 'professional_background']].apply(lambda x: SequenceMatcher(None, 
                                                                                                            x.technical_background.lower() if x.technical_background else '', 
                                                                                                            x.professional_background.lower() if x.professional_background else '')
                                                                                  .ratio(), axis=1),
         lower_i = (df.professional_background + df.technical_background).apply(lambda x: int(' i ' in str(x))))
```

### Transformers

**There is a little bit of "modeling feature engineering" needed.**

1. All text features will need to be one hot encoded
    - The `applicant_location` feature contains _a lot_ of different locations. To avoid the dimensions of the data from getting to out of hand, I'm going to only hot encode locations that have at least 25 observations in the data. All other locations will be set to `"Other"`
2. Full Time Interest contains some data entry messiness where the word "yes" or "no" occasionally repeats, which causes the computer to this "yes yes" and "yes" are different values. Below, I create a transformer that will clean this up. I'm doing it this way in case future observations in production contain the same messiness.
3. I'm creating a `ToDense` transformer because the resulting transformed data is returned as a sparse matrix, but some of the models require the data to be a dense matrix
    - TODO - Set sparse=False on inherited OHE transformers

```python
class EncodeLocation(OneHotEncoder):
    
    locations = ['New York-Northern New Jersey-Long Island, NY-NJ-PA MSA',
                 'Washington-Arlington-Alexandria, DC-VA-MD-WV MSA',
                 'Denver-Aurora, CO MSA',
                 'Houston-Sugar Land-Baytown, TX MSA',
                 'Chicago-Naperville-Joliet, IL-IN-WI MSA',
                 'Austin-Round Rock, TX MSA',
                 'San Francisco-Oakland-Fremont, CA MSA',
                 'Seattle-Tacoma-Bellevue, WA MSA',
                 'Atlanta-Sandy Springs-Marietta, GA MSA',
                 'Los Angeles-Long Beach-Santa Ana, CA MSA',
                 'Dallas-Fort Worth-Arlington, TX MSA',
                 'Philadelphia-Camden-Wilmington, PA-NJ-DE-MD MSA',
                 'Miami-Fort Lauderdale-Pompano Beach, FL MSA',
                 'Detroit-Warren-Livonia, MI MSA',
                 'Colorado Springs, CO MSA',
                 'Phoenix-Mesa-Scottsdale, AZ MSA',
                 'San Diego-Carlsbad-San Marcos, CA MSA',
                 'Tampa-St. Petersburg-Clearwater, FL',
                 'Portland-Vancouver-Beaverton, OR-WA MSA',
                 'Columbus, OH MSA',
                 'Riverside-San Bernardino-Ontario, CA MSA',
                 'Minneapolis-St. Paul-Bloomington, MN-WI MSA',
                 'Las Vegas-Paradise, NV MSA',
                 'Charlotte-Gastonia-Concord, NC-SC MSA',
                 'Orlando-Kissimmee, FL MSA',
                 'Baltimore-Towson, MD MSA',
                 'St. Louis, MO-IL MSA',
                 'Pittsburgh, PA MSA',
                 'Nashville-Davidson-Murfreesboro-Franklin, TN MSA',
                 'Cleveland-Elyria-Mentor, OH MSA',
                 'San Antonio, TX MSA',
                 'Birmingham-Hoover, AL MSA',
                ]
    
    def fit(self, X, y=None):
        X_ = self.replace(X)
        super().fit(X_, y)
        return self
    
    def transform(self, X, y=None):
        X_ = self.replace(X)
        return super().transform(X_)
        
        
    def replace(self, X):
        X_ = X.copy()
        X_[~np.in1d(X_, self.locations)] = 'Other'
        X_[X_ == None] = 'Other'
        return X_
    
class EncodeFullTimeInterest(OneHotEncoder):
    
    r = re.compile('Yes|No')
    
    def fit(self, X, y=None):
        X_ = self.encode(X)
        super().fit(X_)
        return self
        
    def transform(self, X):
        X_ = self.encode(X)
        return super().transform(X_)

    def encode(self, X):
        missing = 'NULL'
        def find_yes_no(x):
            if not x:
                return missing
            match = self.r.findall(x)
            if not match:
                return missing
            return match[0]
        vmatch = np.vectorize(find_yes_no, otypes=[str])
        X_ = vmatch(X) 
        return X_
    
class ToDense(BaseEstimator, TransformerMixin):
    
    def fit(self, X, y):
        return self
    
    def transform(self, X, y=None):
        return X.toarray()
```

```python
from pandas._libs.interval import Interval
def assign_score(probability, interval=False):
    bands = [
        Interval(left=0, right=0.0488),
        Interval(left=0.0488, right=0.0898),
        Interval(left=0.0898, right=0.154),
        Interval(left=0.154, right=0.237),
        Interval(left=0.237, right=1)
    ]
    
    for idx, band in enumerate(bands, start=1):
        if probability > band.left and probability <= band.right:
            if interval:
                return band
            return idx

fig, ax = plt.subplots(figsize=(15,6))
X_ = X.assign(preds=best_model.predict_proba(X)[:,1],
              pred_group=lambda x: x.preds.apply(assign_score, interval=True))
groups = sorted(X_.pred_group.unique())
for idx, group in enumerate(groups):
    target = X_.assign(target=y)[X_.pred_group==group].target
    samples = [np.random.choice(target, size=target.shape[0]).mean() for sample in range(1000)]
    ax.hist(samples, label=f'Group: {idx + 1} |' + f' {round(group.left, 2) * 100}%-{round(round(group.right - .01, 2) * 100, 2)}%'.ljust(12) + f'| Expected Mean: {round(np.mean(samples), 2)}')
    ax.legend()
    # ax.set_title(' Groups')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
```

```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import matplotlib.ticker as mtick
import numpy as np
from pandas._libs.interval import Interval

groups = [
    Interval(left=0, right=0.0483),
    Interval(left=0.0483, right=0.088),
    Interval(left=0.088, right=0.149),
    Interval(left=0.149, right=0.224),
    Interval(left=0.224, right=1)
]

performance_df = pd.DataFrame()
for idx, group in enumerate(groups):
    fig = plt.figure(figsize=(10,6))

    gs = GridSpec(2, 1, figure=fig, height_ratios=[1,3])


    hist_ax = fig.add_subplot(gs[0])
    table_ax = fig.add_subplot(gs[1])
    
    frame = df.assign(preds=best_model.predict_proba(df)[:,1]).query(f'preds > {group.left} and preds <= {group.right}')
    samples = [np.random.choice(frame.target, size=frame.target.shape[0]).mean() for sample in range(300)]
    hist_ax.hist(samples)
    
    xaxis = hist_ax.get_xticks()
    yaxis = hist_ax.get_yticks()
    hist_ax.set_xticks([])
    hist_ax.set_yticks([])
    hist_ax.tick_params(axis='x',which='both', bottom=False,top=False,labelbottom=False)
    table_ax.tick_params(axis='x',which='both', bottom=False,top=False,labelbottom=False)
    hist_ax.tick_params(axis='y',which='both', bottom=False,top=False,labelbottom=False)
    table_ax.tick_params(axis='y',which='both', left=False)
    hist_ax.annotate('{:,.3%}'.format(xaxis[1]), (xaxis[1], yaxis.mean()), color='white')
    hist_ax.annotate('{:,.3%}'.format(np.median(xaxis)), (np.median(xaxis), yaxis.mean()), color='white')
    hist_ax.annotate('{:,.3%}'.format(xaxis[-2]), (xaxis[-2], yaxis.mean()), color='white')
    
    table = (frame.set_index('application_submission_date')
                  .groupby(pd.Grouper(freq='1M'))
                  .agg([np.mean, len])
                  .target
                  .reset_index()
                  .assign(application_submission_date=lambda x: x.application_submission_date.dt.strftime('%m-%d-%Y'))
                  .rename({'application_submission_date':'One Month Period',
                               'mean':'Matriculation Rate',
                               'len':'Sample Size'}, axis=1)
                  .assign(group=idx+1,
                          expected_rate=np.mean(samples) - np.std(samples)))
    performance_df = pd.concat([performance_df, table.rename(lambda x: x.lower().replace(' ', '_'), axis=1).assign(one_month_period=lambda x: pd.to_datetime(x.one_month_period))])
    cell_text = []
    for row in range(len(table)):
        cell_text.append(table.drop(['group', 'expected_rate'], axis=1).iloc[row])
        
    cell_colors = [['white', 'green' if x >= np.min(samples) else 'red', 'white'] for x in table['Matriculation Rate']]
        
    table_viz = table_ax.table(cellText=cell_text, 
                               cellColours=cell_colors, 
                               colLabels=table.columns[:-2], 
                               loc=None)
    table_props = table_viz.properties()
    table_cells = table_props['children']
    for cell in table_cells: 
        # cell.get_text().set_fontsize(20)
        if '.' in cell.get_text()._text:
            cell.get_text().set_color('white')
        else:
            cell.get_text().set_color('black')
            
    plt.setp(table_ax.get_xticklabels(), visible=False)
    plt.setp(table_ax.get_yticklabels(), visible=False)
    table_ax.spines['top'].set_visible(False)
    table_ax.spines['right'].set_visible(False)
    table_ax.spines['left'].set_visible(False)
    table_ax.spines['bottom'].set_visible(False)
    hist_ax.spines['top'].set_visible(False)
    hist_ax.spines['right'].set_visible(False)
    hist_ax.spines['left'].set_visible(False)
    hist_ax.spines['bottom'].set_visible(False)
    table_viz.scale(1,1.5)
    hist_ax.set_title('Group: {}'.format(idx + 1))
    plt.show()
    
```

### Residuals

```python
from matplotlib.lines import Line2D
from matplotlib.patches import Patch

legend_elements = [Line2D([0], [0], color='white', label='Mean Error', linestyle='--'),
                   Line2D([0], [0], color='white', label='Median Error'),
                   Line2D([0], [0], marker='o', markeredgecolor='black', color='green', label='Matriculated', markersize=15),
                   Line2D([0], [0], marker='o', color='red', markeredgecolor='black', label='Did not matriculate', markersize=15)]

def color_logic(x):
    if x < 0:
        color = 'green'
    elif x > 0:
        color = 'red'
    else:
        color = 'yellow'
        
    return color
fig, ax = plt.subplots(figsize=(15,6))
residuals = best_model.predict_proba(X)[:,1] - y
plt.scatter([x for x in range(y.shape[0])], 
            residuals, 
            c=[color_logic(x) for x in residuals.values],
            alpha=.8, edgecolors='black')
neg_class_error = residuals[residuals > 0]
pos_class_error = residuals[residuals < 0]
neg_class_mean_error = neg_class_error.mean()
neg_class_median_error = neg_class_error.median()
pos_class_mean_error = pos_class_error.mean()
pos_class_median_error = pos_class_error.median()
plt.hlines(neg_class_mean_error, -1000, 27000, color='white', linestyle='--', label='Mean Error')
plt.hlines(neg_class_median_error, -1000, 27000, color='white', label='Median Error')
plt.hlines(pos_class_mean_error, -1000, 27000, color='white', linestyle='--')
plt.hlines(pos_class_median_error, -1000, 27000, color='white')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
plt.legend(handles=legend_elements);
```
